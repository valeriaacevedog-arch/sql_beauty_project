```python
cols_to_drop = [
    'variation_desc',
    'value_price_usd',
    'sale_price_usd',
    'child_max_price',
    'child_min_price'
]

df_sephora = df_sephora.drop(columns=cols_to_drop)
print('Columns remaining:', df_sephora.shape[1])
```

```python
df_sephora = df_sephora.dropna(subset=['secondary_category'])
```

```python
df_sephora['reviews'] = df_sephora['reviews'].astype('Int64')
df_sephora['reviews'].dtype
```

```python
df_sephora[bool_cols] = df_sephora[bool_cols].astype(bool)

print(df_sephora[bool_cols].dtypes)
```

```python
#fill with median for  realistic values
rating  = df_sephora["rating"].median()
reviews = df_sephora["reviews"].median()
```

```python
df_sephora["rating"]  = df_sephora["rating"].fillna(rating)
df_sephora["reviews"] = df_sephora["reviews"].fillna(reviews)
```

```python
df_sephora["highlights"] = df_sephora["highlights"].fillna("[]")
```

```python
df_sephora["price_usd"]   = df_sephora["price_usd"].astype(float).round(2)
df_sephora["rating"]      = df_sephora["rating"].astype(float).round(2)
df_sephora["reviews"] = df_sephora["reviews"].round(0).astype(int)
```

```python
# Keep only the columns we need for the analysis
cols = [
    "product_id",
    "product_name",
    "price_usd", 
    "rating",
    "reviews", 
    "primary_category"
]
```

```python
df_sephora = df_sephora[cols].copy()
```

```python
print( {df_sephora.shape})
print(f"{df_sephora.isnull().sum()}")
print(f"{df_sephora.dtypes}")
print(f"{df_sephora.head(3)}")
```

```python
sephora_cats = sorted(df_sephora["primary_category"].unique())
```

```python
dim_category = pd.DataFrame({
    "category_id"  : range(1, len(sephora_cats) + 1),
    "category_name": sephora_cats
})
```

```python
dim_category.to_csv("dim_category.csv", index=False)
```

```python
category_map = dict(zip(dim_category["category_name"], dim_category["category_id"]))
```

```python
df_sephora["category_id"] = df_sephora["primary_category"].map(category_map)
```

```python
sephora_products = df_sephora[[
    "product_id",
    "product_name",
    "price_usd",
    "rating",
    "reviews",
    "category_id",
]]
```

```python
sephora_products.to_csv("sephora_clean.csv", index=False)
```

```python
import pandas as pd
import regex as re

# Step 1 — Audit
special_characters = ['"', "'", '-', '_', '+', '&', '.', '*','%','°','/','’','|',"#"]

for char in special_characters:
    total = 0
    total += df_sephora['product_name'].astype(str).str.count(re.escape(char)).sum()
    print(f"'{char}' → {total} occurrences")
```

```python
for char in special_characters:
    df_sephora['product_name'] = df_sephora['product_name'].str.replace(char, '', regex=False)
```

```python

df_sephora.to_csv('sephora_clean.csv', encoding='utf-8', index=False)
```
