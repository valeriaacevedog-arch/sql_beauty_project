```python
EU_COUNTRIES = {
    "Austria", "Belgium", "Bulgaria", "Croatia", "Cyprus",
    "Czech Republic", "Denmark", "Estonia", "Finland", "France",
    "Germany", "Greece", "Hungary", "Ireland", "Italy", "Latvia",
    "Lithuania", "Luxembourg", "Malta", "Netherlands", "Poland",
    "Portugal", "Romania", "Slovakia", "Slovenia", "Spain", "Sweden"
}
```

```python
countries   = sorted(df_world["country_of_origin"].unique())
dim_country = pd.DataFrame({
    "country_id"  : range(1, len(countries) + 1),
    "country_name": countries,
    "is_eu"       : [1 if c in EU_COUNTRIES else 0 for c in countries]
})
```

```python
country_map = dict(zip(dim_country["country_name"], dim_country["country_id"]))
```

```python
df_world["country_id"] = df_world["country_of_origin"].map(country_map)
```

```python
print("dim_country_id:")
dim_country.to_csv("dim_country_id.csv", index=False)
```

```python

all_genders = sorted(
    set(df_world["gender_target"].unique()) 
)

dim_gender = pd.DataFrame({
    "gender_id"   : range(1, len(all_genders) + 1),
    "gender_label": all_genders
})
```

```python
print("dim_gender:")
print(dim_gender)
dim_gender.to_csv("dim_gender.csv", index=False)
```

```python
gender_map = dict(zip(dim_gender["gender_label"], dim_gender["gender_id"]))
```

```python
df_world["gender_id"]  = df_world["gender_target"].map(gender_map)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.DataFrame({
    'Region': ['EU', 'Non-EU'],
    'Cruelty-Free Products': [5586, 2733]
})

colors = ['#E0517F', '#D0D0D0']  # azul para EU, naranja para Non-EU


plt.figure(figsize=(7,7))
bars = plt.bar(df['Region'], df['Cruelty-Free Products'], color=colors, width = 0.70)


plt.title('Cruelty-Free Products by Region', fontsize=12, fontweight='bold')
plt.xlabel('Region', fontsize=11)
plt.ylabel('Number of Products', fontsize=11)


for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 100, int(yval), ha='center', va='bottom', fontsize=10)


plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.DataFrame({
    'Price Range': ['Under $25', '$25 to $50', '$50 to $100', '$100 to $200'],
    'Avg Rating': [3.02, 2.98, 2.97, 3.01]
})


colors = plt.cm.viridis(range(len(df)))


plt.figure(figsize=(6,4))
bars = plt.bar(df['Price Range'], df['Avg Rating'], color=colors,width = 0.70)


plt.title('Average Rating by Price Range', fontsize=14, fontweight='bold')
plt.xlabel('Price Range')
plt.ylabel('Average Rating')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.005, round(yval, 2),
             ha='center', va='bottom')

plt.ylim(2.9, 3.05)  

plt.tight_layout()
plt.show()
```
